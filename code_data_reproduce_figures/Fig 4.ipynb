# Load Uncertainty Analysis Data

```python
#Load Uncertainty Analysis Data
data = np.load('./output/uncertainty_results.npz', allow_pickle=True)
sediment_list = data['sediment_list'].item()
cost_list = data['cost_list'].item()
install_list= data['install_list'].item()



# Load Scenario Metadata
file_path = "./output/uncertainty_all_scenarios.xlsx"
sheet_num = 0  # Sheet index
df = pd.read_excel(file_path, sheet_name=sheet_num, index_col=0)
matching_columns = ["demand",  "fuel price", "inflow", 
                        "technology upper bound", "technology fix cost", "transline fix cost"]

def find_difference(solution_pair):
    print(solution_pair)
    base_solution  = solution_pair[0]
    develop_solution = solution_pair[1]
    df_1 = df[(df["carbon"] == base_solution) & (df['discount factor'] == 'discount factor medium.xlsx')].copy()
    df_2 = df[(df["carbon"] == develop_solution) & (df['discount factor'] == 'discount factor medium.xlsx')].copy()
    # Store indices before merging
    df_1["index_1"] = df_1.index
    df_2["index_2"] = df_2.index

    # Perform an inner merge on the selected columns to find matching tuples
    matched_pairs = df_1.merge(df_2, on=matching_columns, suffixes=("_1", "_2"))

    # Extract the index pairs
    matching_indices = list(zip(matched_pairs["index_1"], matched_pairs["index_2"]))


    # Print or use the results
    print(f"Number of matched pairs: {len(matching_indices)}")
    
    
    # Calculate cost differences (absolute and relative) for each pair
    cost_differences = {}
    relative_cost_differences = {}
    wind_differences = {}
    relative_wind_differences = {}
    solar_differences = {}
    relative_solar_differences = {}
    storage_differences = {}
    relative_storage_differences = {}
    cost_eco_abs_diff_1={}
    cost_eco_abs_diff_2={}
    cost_eco_abs_diff_3={}
    relative_cost_eco_differences_1={}
    relative_cost_eco_differences_2={}
    relative_cost_eco_differences_3={}
    # Calculate cost differences for each pair
    
    for pair in matching_indices:
        file_1 = f"mekong_2020_288h_4_years_AD110dams__{pair[0]}.nc"
        file_2 = f"mekong_2020_288h_4_years_AD110dams__{pair[1]}.nc"

        # Absolute difference
        cost_abs_diff = cost_list[develop_solution][file_2] - cost_list[base_solution][file_1]
        cost_differences[(pair[0], pair[1])] = cost_abs_diff
        cost_eco_abs_diff_1[(pair[0], pair[1])] = cost_list[develop_solution][file_2] - cost_list[base_solution][file_1] -20*39.8*30/1000
        cost_eco_abs_diff_2[(pair[0], pair[1])] = cost_list[develop_solution][file_2] - cost_list[base_solution][file_1] -12*39.8*30/1000
        cost_eco_abs_diff_3[(pair[0], pair[1])] = cost_list[develop_solution][file_2] - cost_list[base_solution][file_1] -28*39.8*30/1000
        
        
        wind_abs_dif = install_list[develop_solution][(file_2,"Wind")] - install_list[base_solution][(file_1,"Wind")]
        wind_differences[(pair[0], pair[1])] = wind_abs_dif
        
        solar_abs_dif = install_list[develop_solution][(file_2,"Solar")] - install_list[base_solution][(file_1,"Solar")]
        solar_differences[(pair[0], pair[1])] = solar_abs_dif

        storage_abs_dif = install_list[develop_solution][(file_2,"Li-ion")] - install_list[base_solution][(file_1,"Li-ion")]
        storage_differences[(pair[0], pair[1])] = storage_abs_dif
        
        # Relative difference (percentage)
        cost_base_value = cost_list[base_solution][file_1]  # Using L1_2 as the base for percentage
        if cost_base_value != 0:
            rel_diff = (cost_abs_diff / cost_base_value) * 100  # Convert to percentage
        else:
            rel_diff = 0  # Avoid division by zero
        relative_cost_differences[(pair[0], pair[1])] = rel_diff
        
        
         # Relative difference (percentage)  eco
        cost_base_value = cost_list[base_solution][file_1]  # Using L1_2 as the base for percentage
        if cost_base_value != 0:
            rel_diff = (cost_eco_abs_diff_1[(pair[0], pair[1])] / cost_base_value) * 100  # Convert to percentage
        else:
            rel_diff = 0  # Avoid division by zero
        relative_cost_eco_differences_1[(pair[0], pair[1])] = rel_diff
        if cost_base_value != 0:
            rel_diff = (cost_eco_abs_diff_2[(pair[0], pair[1])] / cost_base_value) * 100  # Convert to percentage
        else:
            rel_diff = 0  # Avoid division by zero
        relative_cost_eco_differences_2[(pair[0], pair[1])] = rel_diff
        if cost_base_value != 0:
            rel_diff = (cost_eco_abs_diff_3[(pair[0], pair[1])] / cost_base_value) * 100  # Convert to percentage
        else:
            rel_diff = 0  # Avoid division by zero
        relative_cost_eco_differences_3[(pair[0], pair[1])] = rel_diff
        
        
        wind_base_value = install_list[develop_solution][(file_2,"Wind")]  # Using L1_2 as the base for percentage
        if wind_base_value != 0:
            rel_diff = (wind_abs_dif / wind_base_value) * 100  # Convert to percentage
        else:
            rel_diff = 0  # Avoid division by zero
        relative_wind_differences[(pair[0], pair[1])] = rel_diff
        
        
        solar_base_value = install_list[base_solution][(file_1,"Solar")]  # Using L1_2 as the base for percentage
        if solar_base_value != 0:
            rel_diff = (solar_abs_dif / solar_base_value) * 100  # Convert to percentage
        else:
            rel_diff = 0  # Avoid division by zero
        relative_solar_differences[(pair[0], pair[1])] = rel_diff
        
        storage_base_value = install_list[base_solution][(file_1,"Li-ion")]  # Using L1_2 as the base for percentage
        if solar_base_value != 0:
            rel_diff = (storage_abs_dif / storage_base_value) * 100  # Convert to percentage
        else:
            rel_diff = 0  # Avoid division by zero
        relative_storage_differences[(pair[0], pair[1])] = rel_diff        
    
    
    cost_diff = np.mean(list(cost_differences.values()))
    std_cost_diff = np.std(list(cost_differences.values())) 
    relative_cost_diff = np.mean(list(relative_cost_differences.values()))
    std_relative_cost_differences =  np.std(list(relative_cost_differences.values())) 
    
    cost_diff_eco_1 = np.mean(list(cost_eco_abs_diff_1.values()))
    cost_diff_eco_2 = np.mean(list(cost_eco_abs_diff_2.values()))
    cost_diff_eco_3 = np.mean(list(cost_eco_abs_diff_3.values()))
    combined_values  = list(cost_eco_abs_diff_1.values()) + list(cost_eco_abs_diff_2.values()) + list(cost_eco_abs_diff_3.values())
    std_cost_diff_eco = np.std(combined_values) 
    relative_cost_diff_1 = np.mean(list(relative_cost_eco_differences_1.values())) 
    relative_cost_diff_2 = np.mean(list(relative_cost_eco_differences_2.values())) 
    relative_cost_diff_3 = np.mean(list(relative_cost_eco_differences_3.values())) 
    combined_values  = list(relative_cost_eco_differences_1.values()) + list(relative_cost_eco_differences_2.values()) + list(relative_cost_eco_differences_3.values())
    std_relative_cost_diff_eco = np.std(combined_values) 
    
    wind_diff = np.mean(list(wind_differences.values()))
    std_wind_diff = np.std(list(wind_differences.values())) 
    relative_wind_diff = np.mean(list(relative_wind_differences.values())) 
    std_relative_wind_diff = np.std(list(relative_wind_differences.values())) 
    
    solar_diff = np.mean(list(solar_differences.values()))
    std_solar_diff = np.std(list(solar_differences.values())) 
    relative_solar_diff = np.mean(list(relative_solar_differences.values())) 
    std_relative_solar_diff = np.std(list(relative_solar_differences.values())) 
    
    storage_diff = np.mean(list(storage_differences.values()))
    std_storage_diff = np.std(list(storage_differences.values())) 
    relative_storage_diff = np.mean(list(relative_storage_differences.values()))
    std_relative_storage_diff = np.std(list(relative_storage_differences.values())) 
    
    
    
    
    return [cost_diff, std_cost_diff, relative_cost_diff, 
            wind_diff, std_wind_diff, relative_wind_diff,
           solar_diff,std_solar_diff,relative_solar_diff,
            storage_diff,std_storage_diff,relative_storage_diff,
           cost_diff_eco_1,cost_diff_eco_2,cost_diff_eco_3, 
            std_cost_diff_eco,
            relative_cost_diff_1,relative_cost_diff_2,relative_cost_diff_3,
          
           std_relative_cost_differences, std_relative_wind_diff, std_relative_solar_diff, std_relative_storage_diff,
           std_relative_cost_diff_eco]
    



pari_index = {('GHG1','limit1'):('solution_F2.xlsx', 'solution_L2.xlsx'),
                 ('GHG1','limit2'):('solution_F3.xlsx', 'solution_L3.xlsx'),
                 ('GHG1','limit4'):('solution_F4.xlsx', 'solution_L4_2.xlsx'),
                 ('GHG1','limit6'):('solution_F0_2.xlsx', 'solution_L1_2.xlsx'),
                 ('GHG2','limit1'):('solution_F7.xlsx', 'solution_L7.xlsx'),
                 ('GHG2','limit2'):('solution_F8.xlsx', 'solution_L8.xlsx'),
                 ('GHG2','limit4'):('solution_F9.xlsx', 'solution_L9.xlsx'),
                 ('GHG2','limit6'):('solution_F6.xlsx', 'solution_L6_2.xlsx'),
                 ('GHG3','limit1'):('solution_F10.xlsx', 'solution_L10.xlsx'),
                 ('GHG3','limit2'):('solution_F11.xlsx', 'solution_L11.xlsx'),
                 ('GHG3','limit4'):('solution_F12.xlsx', 'solution_L12.xlsx'),
                 ('GHG3','limit6'):('solution_F5.xlsx', 'solution_L5_2.xlsx'),
                 ('GHG4','limit1'):('solution_F13.xlsx', 'solution_L13.xlsx'),
                 ('GHG4','limit2'):('solution_F14.xlsx', 'solution_L14.xlsx'),
                 ('GHG4','limit4'):('solution_F15.xlsx', 'solution_L15.xlsx'),
                 ('GHG4','limit6'):('solution_F16.xlsx', 'solution_L16.xlsx'),
                 }    
limit_Scenario=['limit1','limit2','limit4','limit6']
GHG_scenario= ['GHG1','GHG2','GHG3','GHG4']
result = {}
for i in limit_Scenario:
    for j in GHG_scenario:
        
        result[(j,i)] = find_difference(pari_index[(j,i)])
```

# prepare data for country-level analysis 

```python


def get_cost_country(data, country):
    ix = list(data.zone.values).index(country)
    cost = (
        data.cost_var_tech_breakdown.sel(zone=country).sum().item() +
        data.cost_fuel_breakdown.sel(zone=country).sum().item() +
        data.cost_fix_tech_breakdown.sel(zone=country).sum().item() +
        data.cost_newtech_breakdown.sel(zone=country).sum().item() +
        data.cost_var_line_breakdown.values[ix, :, :].sum() +
        data.cost_fix_line_breakdown.values[ix, :, :].sum() +
        data.cost_newline_breakdown.values[ix, :, :].sum() +
        data.cost_import_breakdown.sel(zone=country).sum().item()
    )
    return cost


limit_scenarios = ['limit1', 'limit2', 'limit4', 'limit6']
ghg_scenarios = ['GHG1', 'GHG2', 'GHG3', 'GHG4']
solutions = [(g, l, i) for l in limit_scenarios for g in ghg_scenarios for i in [27, 48]]

# Load Output Files for Each Scenario
data_out_Country = {}
for (ghg, limit, run_id) in solutions:
    file_path = f'./output/country_level/mekong_2020_288h_4_years_AD110dams_{ghg}_{limit}_{run_id}_breakdown.nc'
    data_out_Country[(ghg, limit, run_id)] = xr.open_dataset(file_path)

     
# Calculate Country-Level Cost Differences    
countries = ['Cambodia', 'Laos', 'Myanmar', 'Thailand', 'Vietnam']
cost_dif_countries = []

for i in range(16):
    cost_diff = {}
    for country in countries:
        high = get_cost_country(data_out_Country[solutions[i * 2]], country) / 1e9  # in billion CNY
        low = get_cost_country(data_out_Country[solutions[i * 2 + 1]], country) / 1e9
        cost_diff[country] = high - low
    cost_dif_countries.append(cost_diff)

    
# Organize Data for Stacked Bar Plotting
organized_data = []

# Iterate through each dictionary in cost_dif_countries
for entry in cost_dif_countries:
    # Initialize dictionaries to store negative and positive values
    neg_values_dict = {}
    pos_values_dict = {}
    
    # Iterate through key-value pairs in the current dictionary
    for country, value in entry.items():
        if value < 0:
            neg_values_dict[country] = value
        else:
            pos_values_dict[country] = value
    
    # Append the dictionaries to organized_data
    organized_data.append([pos_values_dict, neg_values_dict])

# Extract countries and values
countries = list(cost_dif_countries[0].keys())
```
